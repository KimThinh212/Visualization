```python
import pandas as pd
import urllib.parse # Add this import

# Sử dụng API của NYC Open Data để kéo trực tiếp dữ liệu từ 01/01/2024 đến nay.
# (Mình set limit=300000 vì trung bình NYC có khoảng 100,000 vụ tai nạn mỗi năm)

base_url = "https://data.cityofnewyork.us/resource/h9gi-nx95.csv"
limit = 300000
where_clause_condition = "crash_date >= '2024-01-01T00:00:00'"
encoded_where_clause = urllib.parse.quote_plus(where_clause_condition)

api_url = f"{base_url}?$limit={limit}&$where={encoded_where_clause}"

print("Đang tải dữ liệu 2024-2025 từ NYC Cảnh sát... (Vui lòng đợi 1-2 phút tùy tốc độ mạng)")
df_24_25 = pd.read_csv(api_url)

# Chuyển đổi cột ngày sang kiểu Datetime
df_24_25['crash_date'] = pd.to_datetime(df_24_25['crash_date'])

print("Tải thành công!")
print(f"Kích thước dữ liệu: {df_24_25.shape}")
print(f"Dữ liệu cập nhật từ: {df_24_25['crash_date'].min().date()} đến mới nhất là {df_24_25['crash_date'].max().date()}")

# Hiển thị 3 dòng đầu
display(df_24_25.head(3))
```

```python
# Lọc dữ liệu chỉ lấy từ 01/01/2024 đến 31/12/2025
df_filtered = df_24_25[(df_24_25['crash_date'] >= '2024-01-01') & (df_24_25['crash_date'] <= '2025-12-31')]

# Lưu thành file CSV
file_name = "nyc_crashes_2024_2025.csv"
df_filtered.to_csv(file_name, index=False)

print(f"Đã lọc dữ liệu. Số dòng còn lại: {len(df_filtered)}")
print(f"Đã lưu file thành công: {file_name}")

# Kiểm tra lại khoảng thời gian
print(f"Khoảng thời gian mới: {df_filtered['crash_date'].min().date()} đến {df_filtered['crash_date'].max().date()}")
```

```python
# ==========================================
# STEP 7: HARMONIZING THE NEW API DATA (2024-2025)
# ==========================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define mapping dictionary for the API's specific column names
api_rename_dict = {
    'collision_id': 'Collision ID',
    'crash_date': 'Date',
    'crash_time': 'Time',
    'borough': 'Borough',
    'on_street_name': 'Street Name',
    'cross_street_name': 'Cross Street',
    'latitude': 'Latitude',
    'longitude': 'Longitude',
    'contributing_factor_vehicle_1': 'Contributing Factor',
    'vehicle_type_code1': 'Vehicle Type',
    'vehicle_type_code_1': 'Vehicle Type',
    'number_of_persons_injured': 'Persons Injured',
    'number_of_persons_killed': 'Persons Killed',
    'number_of_pedestrians_injured': 'Pedestrians Injured',
    'number_of_pedestrians_killed': 'Pedestrians Killed',
    'number_of_cyclist_injured': 'Cyclists Injured',
    'number_of_cyclist_killed': 'Cyclists Killed',
    'number_of_motorist_injured': 'Motorists Injured',
    'number_of_motorist_killed': 'Motorists Killed'
}

# Rename the columns
df_24_25_renamed = df_24_25.rename(columns=api_rename_dict)

# 2. Extract only our standard 18 target columns
target_columns = [
    'Collision ID', 'Date', 'Time', 'Borough', 'Street Name', 'Cross Street',
    'Latitude', 'Longitude', 'Contributing Factor', 'Vehicle Type',
    'Persons Injured', 'Persons Killed', 'Pedestrians Injured', 'Pedestrians Killed',
    'Cyclists Injured', 'Cyclists Killed', 'Motorists Injured', 'Motorists Killed'
]

# Ensure no columns are missing
for col in target_columns:
    if col not in df_24_25_renamed.columns:
        df_24_25_renamed[col] = pd.NA

df_24_25_clean = df_24_25_renamed[target_columns].copy()

# 3. Clean Missing Values & Standardize Text
text_cols = ['Borough', 'Street Name', 'Cross Street', 'Contributing Factor', 'Vehicle Type']
df_24_25_clean[text_cols] = df_24_25_clean[text_cols].fillna('Unspecified')
df_24_25_clean['Borough'] = df_24_25_clean['Borough'].astype(str).str.strip().str.title()

# 4. Clean Geographical Data
df_24_25_clean['Latitude'] = pd.to_numeric(df_24_25_clean['Latitude'], errors='coerce')
df_24_25_clean['Longitude'] = pd.to_numeric(df_24_25_clean['Longitude'], errors='coerce')
df_24_25_clean = df_24_25_clean.dropna(subset=['Latitude', 'Longitude'])
df_24_25_clean = df_24_25_clean[(df_24_25_clean['Latitude'] != 0) & (df_24_25_clean['Longitude'] != 0)]

print("✅ 2024-2025 Data Harmonization Complete!")

# ==========================================
# STEP 8: COMBINE INTO A GRAND MASTER DATASET
# ==========================================
# FIXED: Added check for df_master to avoid NameError
if 'df_master' in locals() or 'df_master' in globals():
    df_grand_master = pd.concat([df_master, df_24_25_clean], ignore_index=True)
    print("📊 Combined with historical data (2020-Present)")
else:
    df_grand_master = df_24_25_clean.copy()
    print("⚠️ df_master not found. Using only 2024-2025 data for analysis.")

df_grand_master['Date'] = pd.to_datetime(df_grand_master['Date'])

# ==========================================
# STEP 9: VISUALIZATIONS & ANALYSIS
# ==========================================
sns.set_theme(style="whitegrid")
df_grand_master['Year-Month'] = df_grand_master['Date'].dt.to_period('M')
monthly_grand = df_grand_master.groupby('Year-Month').agg(Total_Collisions=('Collision ID', 'count')).reset_index()
monthly_grand['Date'] = monthly_grand['Year-Month'].dt.to_timestamp()

plt.figure(figsize=(15, 6))
sns.lineplot(data=monthly_grand, x='Date', y='Total_Collisions', color='darkviolet', marker='o')
plt.title('Complete NYC Collision Trend', fontsize=16, fontweight='bold')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập theme cho biểu đồ
sns.set_theme(style='whitegrid')
fig, axes = plt.subplots(2, 1, figsize=(12, 14))

# 1. Phân tích Nguyên nhân dẫn đến tai nạn (Top 10)
# Loại bỏ 'Unspecified' để thấy rõ các nguyên nhân cụ thể
contributing_factors = df_grand_master[df_grand_master['Contributing Factor'] != 'Unspecified']['Contributing Factor'].value_counts().head(10)

sns.barplot(x=contributing_factors.values, y=contributing_factors.index, ax=axes[0], palette='viridis')
axes[0].set_title('Top 10 Nguyên nhân gây tai nạn hàng đầu (2024-2025)', fontsize=15, fontweight='bold')
axes[0].set_xlabel('Số vụ tai nạn')

# 2. Phân tích Loại phương tiện liên quan (Top 10)
vehicle_types = df_grand_master[df_grand_master['Vehicle Type'] != 'Unspecified']['Vehicle Type'].value_counts().head(10)

sns.barplot(x=vehicle_types.values, y=vehicle_types.index, ax=axes[1], palette='magma')
axes[1].set_title('Top 10 Loại phương tiện gặp tai nạn nhiều nhất (2024-2025)', fontsize=15, fontweight='bold')
axes[1].set_xlabel('Số vụ tai nạn')

plt.tight_layout()
plt.show()
```
